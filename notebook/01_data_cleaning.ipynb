9. FIXING duration COLUMN

```python
df['duration'].head()
```

```python
df['duration'].isna().sum()
```

```python
df['duration'].value_counts()
```

```python
df['duration'].str.strip()
```

```python
#We will split seasons and mins into seperate columns.
df['duration_mins'] = df['duration'].str.extract(r"(\d+)").astype(float)
df['dration_seasons']=df['duration'].str.extract(r"(min|Season|Seasons)", expand=False)
```

```python
df = df.rename(columns={"duration_mins": "duration_value"})
df = df.rename(columns={"dration_seasons": "duration_unit"})
```

```python
df["duration_unit"] = df["duration_unit"].replace(
    {"Seasons": "Season"}
)
```

```python
df.head()
```

```python
#Drop duration column
df = df.drop(columns=["duration"])
```
